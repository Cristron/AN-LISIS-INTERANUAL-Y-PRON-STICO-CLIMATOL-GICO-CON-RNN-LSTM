```python
# Encuentra las filas donde la estacion es "Cochabamba Aeropuerto" y crea un nuevo DataFrame
df_cochabamba = df_reset.loc[df_reset['estacion'] == 'Cochabamba_Aeropuerto']
df_cochabamba = df_cochabamba.reset_index(drop=True)

# Hace lo mismo para "2 Sarco"
df_sarco = df_reset.loc[df_reset['estacion'] == 'Sarco']
df_sarco = df_sarco.reset_index(drop=True)
```

```python
null_counts_cochabamba = df_cochabamba.groupby('gestion').apply(lambda x: x.isnull().sum())
null_counts_sarco = df_sarco.groupby('gestion').apply(lambda x: x.isnull().sum())
```

```python
# Supongamos que tienes un DataFrame llamado df_cochabamba
TempMax = df_cochabamba[['Temperatura_Maxima']].copy()
TempMaxNulos = TempMax[TempMax.isnull().any(axis=1)]
TempMaxVecinos = TempMaxNulos.copy()
# Quitar el índice por fecha en TempMaxVecinos
TempMaxVecinos.reset_index(inplace=True)

for indice in TempMaxNulos.index:
    dia_anterior = indice - pd.Timedelta(days=1)
    dia_posterior = indice + pd.Timedelta(days=1)

    # Buscar los valores en el DataFrame TempMax
    valor_dia_anterior = TempMax.at[dia_anterior, "Temperatura_Maxima"]
    valor_dia_posterior = TempMax.at[dia_posterior, "Temperatura_Maxima"]

    # Agregar las filas al DataFrame TempMaxVecinos
    TempMaxVecinos = TempMaxVecinos.append({"Fecha": dia_anterior, "Temperatura_Maxima": valor_dia_anterior}, ignore_index=True)
    TempMaxVecinos = TempMaxVecinos.append({"Fecha": dia_posterior, "Temperatura_Maxima": valor_dia_posterior}, ignore_index=True)

# Ahora TempMaxVecinos contiene las filas con los valores de días anteriores y posteriores a los valores nulos
TempMaxVecinos

```

```python
# Ordenar el DataFrame por la columna 'Fecha'
TempMaxVecinos = TempMaxVecinos.sort_values(by='Fecha')

# Eliminar las fechas repetidas y mantener la primera ocurrencia
TempMaxVecinos = TempMaxVecinos.drop_duplicates(subset='Fecha', keep='first')

# Establecer 'Fecha' como el índice
TempMaxVecinos.set_index('Fecha', inplace=True)

# Establecer como el tipo como float
TempMaxVecinos['Temperatura_Maxima'] = TempMaxVecinos['Temperatura_Maxima'].astype(float)

TempMaxVecinos
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
# Supongamos que df es tu DataFrame, y 'a_rellenar' es la columna que quieres completar
series = df_cochabamba['Temperatura_Maxima']

#Eliminar los valores de nan
series_droppednan = series.dropna()

# Crear secuencias de los datos
sequence_length = 15
sequences = []
for i in range(len(series_droppednan) - sequence_length):
    sequences.append(series_droppednan[i:i+sequence_length].values)
sequences = np.array(sequences)

# Ejemplos y etiquetas
X = sequences[:, :-1]
Y = sequences[:, -1]

# Redimensionar los ejemplos para el LSTM
X = np.expand_dims(X, -1)

# Crear el modelo
model = Sequential()
model.add(LSTM(200, activation='relu', input_shape=(X.shape[1], 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Entrenar el modelo
model.fit(X, Y, epochs=60, verbose=1)

# Utilizar el modelo para introducir valores faltantes
for i in range(len(series)):
    if pd.isnull(series[i]):
        input = np.expand_dims(np.expand_dims(series[i-sequence_length+1:i].values, -1), 0)
        series[i] = model.predict(input)[0, 0]

# Ahora series debería tener los valores faltantes llenados
```

```python
# Supongamos que tienes los DataFrames TempMaxVecinos y TempMaxRNN
# Asegúrate de que ambos tengan la fecha como índice en caso de que no lo tengan
TempMaxRNN = pd.DataFrame(series)
# Supongamos que tienes un DataFrame llamado df y quieres renombrar la columna 'ColumnaOriginal' a 'NuevaColumna'
TempMaxVecinos.rename(columns={'Temperatura_Maxima': 'Original'}, inplace=True)


# Realizar la unión en TempMaxVecinos
TempMaxVecinos = TempMaxVecinos.join(TempMaxRNN, how='left')

# Renombrar la columna resultante como "RNN"
TempMaxVecinos.rename(columns={'Temperatura_Maxima': 'RNN'}, inplace=True)

# Asegúrate de reemplazar 'columna_original_de_RNN' con el nombre real de la columna en TempMaxRNN que deseas agregar

# El DataFrame TempMaxVecinos ahora incluirá la columna de TempMaxRNN bajo el nombre "RNN"
TempMaxVecinos
```
