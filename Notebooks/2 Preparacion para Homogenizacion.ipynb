```python
df1 = pd.read_csv("2000--2004.csv", delimiter=';')
df2 = pd.read_csv("2005--2009.csv", delimiter=';')
df3 = pd.read_csv("2010--2014.csv", delimiter=';')
df4 = pd.read_csv("2015--2019.csv", delimiter=';')
df5 = pd.read_csv("2020--2023.csv", delimiter=';')
```

```python
dataframes = [df1, df2, df3, df4, df5]

# Utiliza la funci√≥n concat() para concatenarlos por filas (eje 0)
df = pd.concat(dataframes, axis=0)
df_reset = df.reset_index()
df_reset = df_reset.drop(columns=['index'])
df_reset
```

```python
print(df_reset.dtypes)
```

```python
# Encuentra las filas donde la estacion es "Cochabamba Aeropuerto" y crea un nuevo DataFrame
df_cochabamba = df_reset.loc[df_reset['estacion'] == 'Cochabamba_Aeropuerto']
df_cochabamba = df_cochabamba.reset_index(drop=True)

# Hace lo mismo para "2 Sarco"
df_sarco = df_reset.loc[df_reset['estacion'] == 'Sarco']
df_sarco = df_sarco.reset_index(drop=True)
```

```python
df_cochabamba = df_cochabamba[df_cochabamba['gestion'] != 2023]
df_sarco = df_sarco[df_sarco['gestion'] != 2023]
```

```python
df_cochabamba['Fecha'] = pd.to_datetime(df_cochabamba['Fecha'], dayfirst=True)
df_cochabamba = df_cochabamba.set_index('Fecha')
df_cochabamba = df_cochabamba.resample('D').asfreq()

df_sarco['Fecha'] = pd.to_datetime(df_sarco['Fecha'], dayfirst=True)
df_sarco = df_sarco.set_index('Fecha')
df_sarco = df_sarco.resample('D').asfreq()
```

```python

columnas_seleccionadas = ['Precipitacion', 'Temperatura_Maxima', 'Temperatura_Minima', 'eTo']
```

```python
datos_c = df_cochabamba[columnas_seleccionadas].copy()
datos_s = df_sarco[columnas_seleccionadas].copy()
```

```python
df_cochabamba = datos_c.copy()
df_sarco = datos_s.copy()
```

```python
# Extraer el d√≠a, el mes y el a√±o en columnas separadas
df_cochabamba['Dia'] = df_cochabamba.index.day
df_cochabamba['Mes'] = df_cochabamba.index.month
df_cochabamba['Gestion'] = df_cochabamba.index.year

df_sarco['Dia'] = df_sarco.index.day
df_sarco['Mes'] = df_sarco.index.month
df_sarco['Gestion'] = df_sarco.index.year
```

```python
# Suponiendo que deseas guardar el DataFrame en un archivo CSV llamado "Aeropuerto.csv"
ruta_archivo = "/content/drive/My Drive/BaseDeDatos/Aeropuerto.txt"

# Guardar el DataFrame en el archivo CSV
df_cochabamba.to_csv(ruta_archivo, index=False, sep=',')
```

```python
# Suponiendo que deseas guardar el DataFrame en un archivo CSV llamado "Sarco.csv"
ruta_archivo = "/content/drive/My Drive/BaseDeDatos/Sarco.txt"

# Guardar el DataFrame en el archivo CSV
df_sarco.to_csv(ruta_archivo, index=False, sep=',')
```

```python
!cp "/content/drive/My Drive/BaseDeDatos/ClimateEnginePre.csv" "/content/"
!cp "/content/drive/My Drive/BaseDeDatos/ClimateEngineTMax.csv" "/content/"
!cp "/content/drive/My Drive/BaseDeDatos/ClimateEngineTmin.csv" "/content/"

Prec = pd.read_csv("ClimateEnginePre.csv",sep=';')
TMax = pd.read_csv("ClimateEngineTMax.csv",sep=';')
Tmin = pd.read_csv("ClimateEngineTmin.csv",sep=';')
```

```python
# Eliminar los nombres de las columnas
Prec.columns = [None] * len(Prec.columns)
TMax.columns = [None] * len(TMax.columns)
Tmin.columns = [None] * len(Tmin.columns)
```

```python
Prec.columns = ['Fecha','Precipitacion']
TMax.columns = ['Fecha','Temperatura_Maxima']
Tmin.columns = ['Fecha','Temperatura_Minima']
```

```python
Prec
```

```python
Prec['Fecha'] = pd.to_datetime(Prec['Fecha'], format='%Y-%m-%d:%I %p')
Prec['Fecha'] = pd.to_datetime(Prec['Fecha'])
Prec.set_index('Fecha', inplace=True)

TMax['Fecha'] = pd.to_datetime(TMax['Fecha'], format='%Y-%m-%d:%I %p')
TMax['Fecha'] = pd.to_datetime(TMax['Fecha'])
TMax.set_index('Fecha', inplace=True)

Tmin['Fecha'] = pd.to_datetime(Tmin['Fecha'], format='%Y-%m-%d:%I %p')
Tmin['Fecha'] = pd.to_datetime(Tmin['Fecha'])
Tmin.set_index('Fecha', inplace=True)
```

```python
Prec['Precipitacion'] = Prec['Precipitacion'].str.replace(',', '.')
Prec['Precipitacion'] = Prec['Precipitacion'].astype(float)

TMax['Temperatura_Maxima'] = TMax['Temperatura_Maxima'].str.replace(',', '.')
TMax['Temperatura_Maxima'] = TMax['Temperatura_Maxima'].astype(float)

Tmin['Temperatura_Minima'] = Tmin['Temperatura_Minima'].str.replace(',', '.')
Tmin['Temperatura_Minima'] = Tmin['Temperatura_Minima'].astype(float)
```

```python
sat = pd.merge(Prec, TMax, on='Fecha', how='outer')
```

```python
sat = pd.merge(sat, Tmin, on='Fecha', how='outer')
```

```python
sat['Nueva_Columna1'] = np.nan
```

```python
# Extraer el d√≠a, el mes y el a√±o en columnas separadas
sat['Dia'] = sat.index.day
sat['Mes'] = sat.index.month
sat['Gestion'] = sat.index.year
```
